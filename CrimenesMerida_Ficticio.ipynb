```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
#Cargar el archivo datos/asaltos_yucatan.csv en un DataFrame llamado data
# si es necesario usar la opcion encoding = "ISO-8859-1"
data = pd.read_csv('datos/asaltos_yucatan.csv', encoding = "ISO-8859-1")
```

```python
#Mostrar las primeras filas del DataFrame data
data.head()
```

```python
mapa=folium.Map(location=[20.974135031910095, -89.62639516740302],zoom_start=9)
```

```python
#Recorremos el dataframe
for crimen in range(len(data)):
    lat=data.iloc[crimen]['lat']
    lon=data.iloc[crimen]['long']
    cat=data.iloc[crimen]['Categoria']
    #Para asignar el logo
    if (data.iloc[crimen]['Categoria']=='asalto'):
        logo='gun'
    else:
        logo='house'
    #Agregamos el marcador para cada crimen
    folium.Marker(location=[lat, lon],popup='<strong>'+ cat+'</strong>',
              tooltip='<strong>Presiona aquí para ver el tipo de crimen</strong>',
              icon=folium.Icon(color='red',prefix='fa',icon=logo)).add_to(mapa)
mapa
    
```
